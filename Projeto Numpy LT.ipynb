```python
import numpy as np
import matplotlib.pyplot as plt
url = 'https://raw.githubusercontent.com/allanspadini/numpy/dados/citrus.csv'
dado = np.loadtxt(url, delimiter=',',usecols=np.arange(1, 6, 1), skiprows=1)
```

```python
dadot = dado.T
dadot
```

```python
diametrolar = dadot[0,0:5000]
diametrolar
```

```python
diamtor = dadot[0,5000:]
diamtor
```

```python
pesolal = dadot[1,0:5000]
pesolal
```

```python
pesotor = dadot[1,5000:]
pesotor
```

```python
plt.plot(pesolal, diametrolar)
```

```python
plt.plot(pesotor, diamtor)
```

```python
## REGRESSÃO LINEAR
# a = coeficiente angular
# n = numero de elementos
# Y = Moscow
# X = datas

#obtendo o coeficiente angular
# a = (n * soma(X*Y) - soma(X) * soma(y)) / (n * soma (X**2) - soma(X)**2)
```

```python
X = pesolal
Y = diametrolar
n = np.size(X)
a = (n* np.sum(X*Y) - np.sum(X)*np.sum(Y)) / (n*np.sum(X**2) - np.sum(X)**2)
```

```python
b = np.mean(Y) - a * np.mean(X)
```

```python
y = a*X+b
np.linalg.norm(diametrolar-y)
```

```python
plt.plot(pesolal, diametrolar)
plt.plot(X, y)
```
